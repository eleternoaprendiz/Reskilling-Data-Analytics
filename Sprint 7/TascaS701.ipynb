Exercici 1
Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.
La funció ha de classificar el resultat en les seves respectives categories

```python
# Definimos la función cálculo_IMC:

def calculo_IMC(peso,altura):
    #Cálculo del IMC:
    IMC = round(peso/pow(altura,2),2)
    #Impresión del IMC:
    print(f"Su índice de masa corporal (IMC) es de {IMC}.")

    #Clasificación según IMC:
    if IMC < 18.5:
        print("Según su IMC, su peso se encuentra por debajo de los valores adecuados a su altura.")
    elif IMC < 25:
        print("Según su IMC, su peso se encuentra en el rango adecuado a su altura.")
    elif IMC < 30:
        print("Según su IMC, su peso se sitúa en el rango del sobrepeso.")
    else:
        print("Según su IMC, su peso se sitúa en el rango de la obesidad.")

# Pedimos a la persona usuaria su peso y altura.
# Controlamos que los datos introducidos sean los que pedimos, int o float.
# Si la persona usuaria introduce otro tipo de datos, saltará la excepción ValueError, donde le indicaremos a la persona
# el formato correcto de los datos y le pedimos que los vuelva a introducir.
# Cuando los datos sean correctos, la variable dato_peso o dato_altura pasará de False a True y se seguirá ejecutando el código.

dato_peso = False

while dato_peso == False:
    try:
        peso_usuario = float(input("Por favor, introduzca su peso en kilogramos (hasta 2 decimales): "))
        dato_peso = True
    except ValueError:
        print("Dato erróneo. Por favor, introduzca de nuevo su peso en kilogramos (hasta 2 decimales). Use solo dígitos y el punto como separador de decimales.")


dato_altura = False

while dato_altura == False:
    try:
        altura_usuario = float(input("Por favor, introduzca su altura en metros (hasta 2 decimales): "))
        dato_altura = True
    except ValueError:
        print("Dato erróneo. Por favor, introduzca de nuevo su altura en metros (hasta 2 decimales). Use solo dígitos y el punto como separador de decimales.")

# Llamamos a la función:

calculo_IMC(peso_usuario,altura_usuario)
```

Exercici 2
Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

```python
# Vamos a crear una función por conversión, que contenga el cálculo
# e imprima el valor en la nueva unidad de temperatura.

# Conversión de temperatura a partir de grados Celsius:
def Celsius_a_Kelvin(temp):
    temp_Kelvin = temp + 273.15
    print(f"Corresponde a {temp_Kelvin} ºK")

def Celsius_a_Fahrenheit(temp):
    temp_Fahrenheit = round(9 * temp / 5 + 32, 2)
    print(f"Corresponde a {temp_Fahrenheit} ºF")

def Celsius_a_Rankine(temp):
    temp_Rankine = round(9 * temp / 5 + 491.67, 2)
    print(f"Corresponde a {temp_Rankine} ºRa")

def Celsius_a_Reaumur(temp):
    temp_Reaumur = round(4 * temp / 5, 2)
    print(f"Corresponde a {temp_Reaumur} ºRe")

# Conversión de temperatura a partir de grados Kelvin:
def Kelvin_a_Celsius(temp):
    temp_Celsius = temp - 273.15
    print(f"Corresponde a {temp_Celsius} ºC")

def Kelvin_a_Fahrenheit(temp):
    temp_Fahrenheit = round(9 * (temp - 273.15) / 5 + 32, 2)
    print(f"Corresponde a {temp_Fahrenheit} ºF") 

def Kelvin_a_Rankine(temp):
    temp_Rankine = round(9 * (temp - 273.15) / 5 + 491.67, 2)
    print(f"Corresponde a {temp_Rankine} ºRa")

def Kelvin_a_Reaumur(temp):
    temp_Reaumur = round(4 * (temp - 273.15) / 5, 2)
    print(f"Corresponde a {temp_Reaumur} ºRe")

# Conversión de temperatura a partir de grados Fahrenheit
def Fahrenheit_a_Celsius(temp):
    temp_Celsius = round((5 * (temp - 32))/9, 2)
    print(f"Corresponde a {temp_Celsius} ºC")

def Fahrenheit_a_Kelvin(temp):
    temp_Kelvin = round((5 * (temp - 32))/9 + 273.15, 2)
    print(f"Corresponde a {temp_Kelvin} ºK")

def Fahrenheit_a_Rankine(temp):
    temp_Rankine = temp + 459.67
    print(f"Corresponde a {temp_Rankine} ºRa")

def Fahrenheit_a_Reaumur(temp):
    temp_Reaumur = round(4 * (temp - 32) / 9, 2)
    print(f"Corresponde a {temp_Reaumur} ºRe")

# Conversión de temperatura a partir de grados Rankine
def Rankine_a_Celsius(temp):
    temp_Celsius = round((5 * (temp - 491.67)) / 9, 2)
    print(f"Corresponde a {temp_Celsius} ºC")

def Rankine_a_Kelvin(temp):
    temp_Kelvin = round((5 * (temp - 491.67)) / 9 + 273.15, 2)
    print(f"Corresponde a {temp_Kelvin} ºK")

def Rankine_a_Fahrenheit(temp):
    temp_Fahrenheit = temp - 459.67
    print(f"Corresponde a {temp_Fahrenheit} ºF")

def Rankine_a_Reaumur(temp):
    temp_Reaumur = round(4 * (temp - 491.67) / 9, 2)
    print(f"Corresponde a {temp_Reaumur} ºRe")

# Conversión de temperatura a partir de grados Réaumur
def Reaumur_a_Celsius(temp):
    temp_Celsius = round(5 * temp / 4, 2)
    print(f"Corresponde a {temp_Celsius} ºC")

def Reaumur_a_Kelvin(temp):
    temp_Kelvin = round(5 * temp / 4 + 273.15, 2)
    print(f"Corresponde a {temp_Kelvin} ºK")

def Reaumur_a_Fahrenheit(temp):
    temp_Fahrenheit = round(9 * temp / 4 + 32, 2)
    print(f"Corresponde a {temp_Fahrenheit} ºF")

def Reaumur_a_Rankine(temp):
    temp_Rankine = round(9 * temp / 4 + 491.67, 2)
    print(f"Corresponde a {temp_Rankine} ºRa")

# Agrupamos estas funciones según la unidad de temperatura
# que nos indique la persona usuaria.

def convertir_Celsius(temp):
    Celsius_a_Kelvin(temp)
    Celsius_a_Fahrenheit(temp)
    Celsius_a_Rankine(temp)
    Celsius_a_Reaumur(temp)

def convertir_Kelvin(temp):
    Kelvin_a_Celsius(temp)
    Kelvin_a_Fahrenheit(temp)
    Kelvin_a_Rankine(temp)
    Kelvin_a_Reaumur(temp)

def convertir_Fahrenheit(temp):
    Fahrenheit_a_Celsius(temp)
    Fahrenheit_a_Kelvin(temp)
    Fahrenheit_a_Rankine(temp)
    Fahrenheit_a_Reaumur(temp)

def convertir_Rankine(temp):
    Rankine_a_Celsius(temp)
    Rankine_a_Kelvin(temp)
    Rankine_a_Fahrenheit(temp)
    Rankine_a_Reaumur(temp)

def convertir_Reaumur(temp):
    Reaumur_a_Celsius(temp)
    Reaumur_a_Kelvin(temp)
    Reaumur_a_Fahrenheit(temp)
    Reaumur_a_Rankine(temp)

# Creamos un diccionario temperaturas para manejar la entrada de las unidades por la persona usuaria:

diccionario_temperaturas = {
    1 : ('Celsius', 'ºC'),
    2 : ('Kelvin', 'ºK'),
    3 : ('Fahrenheit', 'ºF'),
    4 : ('Rankine', 'ºRa'),
    5 : ('Reáumur', 'ºRe'),
}

# Solicitamos a la persona usuaria que introduzca la temperatura. Controlamos que el formato sea correcto;
# de no ser así, avisamos a la persona usuaria a través del manejo del error:

dato_temperatura = False

while dato_temperatura == False:
    try:
        temp_entrada = float(input("Por favor, introduzca el valor de la temperatura: "))
        dato_temperatura = True
    except ValueError:
        print("Dato erróneo. Por favor, introduzca de nuevo el valor de la temperatura. Use solo dígitos y el punto como separador de decimales.")

# Solicitamos a la persona usuaria que escoja una de las opciones de unidad de temperatura.
# Si introduce un valor erróneo, la avisamos para que escoja solo las opciones posibles.

unidad_correcta = False

while unidad_correcta == False:
    try:
        unidad_entrada = int(input("Por favor, indique la unidad de temperatura: \n 1. Celsius  \n 2. Kelvin \n 3. Fahrenheit  \n 4. Rankine  \n 5. Réaumur"))
        if unidad_entrada not in diccionario_temperaturas.keys():
            print("Opción incorrecta. Por favor, vuelva a introducir la unidad de la temperatura.")
        else:
            unidad_correcta = True
    except ValueError:
        print("Valor erróneo. Por favor, introduzca solo números del 1 al 5.")

# Imprimimos el valor y las unidades de la temperatura que nos ha indicado la persona usuaria
# y su correspondencia en el resto de unidades de temperatura:

print(f"La temperatura medida es de {temp_entrada} {diccionario_temperaturas.get(unidad_entrada)[-1]}")
if unidad_entrada == 1:
    convertir_Celsius(temp_entrada)
elif unidad_entrada == 2:
    convertir_Kelvin(temp_entrada)
elif unidad_entrada == 3:
    convertir_Fahrenheit(temp_entrada)
elif unidad_entrada == 4:
    convertir_Rankine(temp_entrada)
elif unidad_entrada == 5:
    convertir_Reaumur(temp_entrada)
else:
    print("Álex, algo ha salido mal y tienes que repetir el código.")
```

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
def contador_palabras(texto):
    # Sustituimos todo caracter que no sea alfabético por un espacio:
    for caracter in texto:
        if caracter.isalpha() == False:
            texto = texto.replace(caracter, ' ')
        texto = texto.lower() #lo bajamos todo a minúsculas

    # Creamos la lista palabras:
    palabras = []

    # Y un diccionario para guardar las veces que aparece cada palabra:
    diccionario_texto = {}

    # Y almacenamos en la lista las palabras del texto
    palabras.extend(texto.split())
    # Ordenamos alfabéticamente la lista
    palabras.sort()

    # Iteramos a lo largo de la lista palabras.
    # Contamos cuántas veces aparece el elemento que estamos iterando
    # y, una vez lo contamos, lo eliminamos de la lista para no repetir el recuento
    # cuando la palabra aparezca más adelante en el texto:
    for i in palabras:
        n = palabras.count(i)
        # print(f'"{i}" aparece {n} veces.')
        for j in palabras:
            if j == i:
                palabras.remove(j)
        # Guardamos en un diccionario los pares palabra y número de apariciones:
        diccionario_texto[i] = n
        
    return(diccionario_texto)
 
# Importamos un texto:
with open('D:\\CiberConnecta_17\\Downloads\\tu_me_quieres_blanca.txt', mode='r', encoding='utf-8') as archivo: #Versión Windows
    contenido = archivo.read()

# Ejecutamos la función:
contador_palabras(contenido)
```

Exercici 4
Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
def diccionario_inverso(diccionario):

    # Creamos una lista para las claves y otra lista para los valores del diccionario de la persona usuaria:
    lista_claves = []
    lista_valores = []

    # Definimos el diccionario inverso que incorporará los valores como claves y viceversa:
    diccionario_inverso = {}

    # Iteramos a lo largo del diccionario suministrado:
    for i in diccionario:
        # Comprobamos si el valor del diccionario se ha incorporado anteriormente a lista_valores;
        # en tal caso, generamos un mensaje de error y salimos de la función
        # (un diccionario, por definición, no tiene claves repetidas, así que no hará falta
        # ejecutar un bucle similar para las claves):
        if diccionario[i] in lista_valores:
            return print("Múltiples valores iguales.") # Salida de la función
        # Si no hay valores repetidos, añadimos clave y valor a sus respectivas listas:
        else:
            lista_claves.append(i)
            lista_valores.append(diccionario[i])

    # Rellenamos el nuevo diccionario intercambiando claves y valores del diccionari suministrado:
    j = 0
    while j < len(lista_claves):
        diccionario_inverso[lista_valores[j]] = lista_claves[j]
        j += 1

    print(diccionario_inverso)

# Creamos un diccionario de prueba:
diccionario_prueba = {'Betis':'1', 'Barça':'3', 'Espanyol':'6', 'Europa':'4', 'El Prat':'5'}

# Y probamos la función:
diccionario_inverso(diccionario_prueba)
```

Exercici 1
Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.

```python
def diccionario_inverso(diccionario):

    # Creamos una lista para las claves y otra lista para los valores del diccionario de entrada:
    lista_claves = []
    lista_valores = []

    # Definimos el diccionario inverso de salida:
    diccionario_inverso = {}

    # Iteramos a lo largo del diccionario
    for i in diccionario:
        # Comprobamos si el valor del diccionario se ha incorporado anteriormente a lista_valores:
        if diccionario[i] in lista_valores:
            posicion = lista_valores.index(diccionario[i])
            # En la primera repetición (que comprobamos porque el elemento en la posición indicada no será una lista)
            # transformamos ese elemento en una lista,
            # así podemos añadir, al salir de este bucle, la clave repetida en la lista
            if type(lista_claves[posicion]) != list:
                lista_claves[posicion] = [lista_claves[posicion]]
            # Y añadimos la(s) clave(s) repetida(s) en la lista
            lista_claves[posicion].append(i)
        else:
            # Si no hay repeticiones, mantenemos el método del ejercicio anterior:
            lista_claves.append(i)
            lista_valores.append(diccionario[i])

    # Rellenamos el nuevo diccionario intercambiando claves y valores:
    j = 0
    while j < len(lista_claves):
        diccionario_inverso[lista_valores[j]] = lista_claves[j]
        j += 1

    print(diccionario_inverso)

# Creamos un diccionario de prueba:
diccionario_prueba = {'Betis':'1', 'Barça':'2', 'Espanyol':'2', 'Europa':'4', 'El Prat':'2', 'Sevilla':'3', 'Athletic': 6, 'Oviedo': 6, 'Sporting': 6}

# Y probamos el funcionamiento:
diccionario_inverso(diccionario_prueba)
```

Exercici 2
Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: **[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]**

```python
def conversor_datos(lista):
    # Definimos las listas que contendrán los datos convertidos a decimales,
    # los que no se pueden convertir y una tercera lista para almacenar
    # los datos extraídos de valores de tipo lista, tupla o diccionario:
    lista_decimales = []
    lista_no_decimales = []
    lista_solo_valores = []

    # Sacamos todos los valores de listas, tuplas y diccionarios a una lista con solo valores:
    for i in lista:
        if type(i) == list or type(i) == tuple:
            lista_solo_valores.extend(i)       
        elif type(i) == dict:
            lista_solo_valores.extend(i)
            lista_solo_valores.extend(i.values())
        else:
            lista_solo_valores.append(i)

    # Ahora aplicamos la conversión de formato a estos valores y los separamos en sendas listas:

    for k in lista_solo_valores:
        try:
            float(k)
            lista_decimales.append(float(k))
        except ValueError:
            lista_no_decimales.append(k)

    print(f"Los valores convertidos a números decimales son {lista_decimales}")
    print(f"Los valores que no se pueden convertir en números decimales son {lista_no_decimales}")

# Tomamos la lista que nos proporciona la persona usuaria
lista_de_cosas = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4'], {'Betis':'1', 'Barça':'2', 'Espanyol':'2', 'Europa':'4', 'El Prat':'2', 'Sevilla':'3', 'Athletic': 3, 'Oviedo': 6, 'Sporting': 6}]

#Y probamos la función:
conversor_datos(lista_de_cosas)
```

Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
def diccionario_contador(texto):

    # Importamos la biblioteca string para usar la función ascii_lowercase
    # y generar fácilmente una lista con el alfabeto (inglés)
    import string 
    
    # Creamos un diccionario con el recuento de palabras del texto:    
    diccionario_texto = contador_palabras(texto)
    # Creamos la variable letra, que recorrerá el abecedario:
    letra = string.ascii_lowercase

    # Y el diccionario qué queremos presentar, cuyos pares serán:
    # clave: letra del abecedario;
    # valor: diccionario con el recuento de palabras, ordenado alfabéticamente,
    # correspondientes a la letra del abecedario:
    diccionario_contador_letra = {}

    # Iteramos por cada letra del abecedario
    for l in letra:
        diccionario_letra = {}
        # Recorremos el diccionario_texto. Por cada pareja clave:valor, nos fijamos si la primera letra de la clave
        # del diccionario_texto coincide con el carácter de la iteración l de la lista letra. Si coincide,
        # incorporamos el par palabra:recuento en diccionario_letra para este valor de l:
        for k in diccionario_texto:
            if k[0] == l:
                diccionario_letra[k] = diccionario_texto.get(k)
        # Si la longitud lista diccionario_letra es mayor que cero, añadiremos al diccionario_contador_letra
        # el par letra:{diccionaro_letra} para el valor l, 
        if len(diccionario_letra) != 0:
            diccionario_contador_letra[l] = diccionario_letra

    return diccionario_contador_letra

# Importamos un texto:
with open('D:\\CiberConnecta_17\\Downloads\\tu_me_quieres_blanca.txt', mode='r', encoding='utf-8') as archivo:
    contenido = archivo.read()

# Y probamos la función:
diccionario_contador(contenido)
```
